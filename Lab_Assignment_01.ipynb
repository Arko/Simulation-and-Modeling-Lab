```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
vector1 = np.random.rand(15)
vector2 = np.random.rand(15)

plt.figure(figsize=(8, 5))
plt.plot(vector1, marker='o', linestyle='-', color='b', label='Vector 1')
plt.plot(vector2, marker='s', linestyle='--', color='r', label='Vector 2')
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Plot of Two Random Vectors")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

matrix1 = np.random.rand(4, 4)

plt.figure(figsize=(6, 5))
sns.heatmap(matrix1, annot=True, cmap='coolwarm', linewidths=0.5, xticklabels=['A', 'B', 'C', 'D'], yticklabels=['W', 'X', 'Y', 'Z'])
plt.title("Heatmap of a 4x4 Random Matrix")
plt.show()
```

```python
matrix2 = np.random.rand(4, 4)
matrix3 = np.random.rand(4, 4)

matrix_add = matrix2 + matrix3
matrix_sub = matrix2 - matrix3
matrix_mult = matrix2 * matrix3
```

```python
plt.figure(figsize=(6, 4))
plt.bar(range(16), matrix_add.flatten(), color='g')
plt.title("Matrix Addition")
plt.xlabel("Element Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.bar(range(16), matrix_sub.flatten(), color='b')
plt.title("Matrix Subtraction")
plt.xlabel("Element Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.bar(range(16), matrix_mult.flatten(), color='r')
plt.title("Matrix Multiplication")
plt.xlabel("Element Index")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
print("Matrix Addition: Increases the overall values as elements are summed.")
print("Matrix Subtraction: Highlights the differences between the two matrices.")
print("Matrix Multiplication: Element-wise multiplication can amplify large values and suppress small values.")
```
