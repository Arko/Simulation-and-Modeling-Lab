```python
import numpy as np
from scipy.stats import kstest, norm
import matplotlib.pyplot as plt
```

1. **Load the Dataset and:**

---



```python
temperature_data = np.array([75, 85, 70, 82, 90, 77, 65, 88, 74, 79])
```

2. **Extract Temperature Scores**:

---



```python
print("Daily Temperatures:", temperature_data)
```

3. **Center and Scale the Data**:

---



```python
mean_val = 80
std_dev = 5

scaled_data = (temperature_data - mean_val) / std_dev
```

4. **Perform the K-S Test:**

---



```python
ks_result = kstest(scaled_data, 'norm')
```

5. **Output the Results**:

---



```python
print("\n--- K-S Test Results ---")
print(f"Test decision (h): {int(ks_result.pvalue < 0.05)}")
print(f"p-value: {ks_result.pvalue:.4f}")
print(f"KS statistic: {ks_result.statistic:.4f}")
```

6. **Plot the Empirical CDF**:

---



```python
sorted_data = np.sort(scaled_data)
ecdf_y = np.arange(1, len(scaled_data)+1) / len(scaled_data)
normal_cdf = norm.cdf(sorted_data)

plt.figure(figsize=(8, 5))
plt.plot(sorted_data, ecdf_y, label='Empirical CDF', color='blue', linewidth=2)
plt.plot(sorted_data, normal_cdf, label='Standard Normal CDF', color='red', linestyle='--', linewidth=2)
plt.title("K-S Test: Empirical CDF vs Normal CDF")
plt.xlabel("Standardized Temperature Scores")
plt.ylabel("Cumulative Probability")
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

7. **Write a Conclusion**:

---



**Conclusion**

---



The One-Sample KolmogorovSmirnov (K-S) Test was used to evaluate whether the 10 days of temperature data follow a normal distribution.

- **Test decision (h):** 0  
- **p-value:** ~0.7547  
- **KS statistic:** ~0.1835  

Because the p-value is greater than the 0.05 threshold, we **fail to reject the null hypothesis**. This means there is **no statistical evidence** to suggest that the temperature data does **not** follow a normal distribution. Therefore, it is reasonable to conclude that the sample temperatures could have come from a normally distributed population.
