```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Generate two vectors with 15 random floats and plot them
x = np.arange(15)
y1 = np.random.rand(15)
y2 = np.random.rand(15)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, marker='o', label='Vector 1')
plt.plot(x, y2, marker='s', label='Vector 2')
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Random Vectors Plot")
plt.legend()
plt.grid()
plt.show()

# Task 2: Create a 4x4 matrix and visualize as heatmap
matrix = np.random.rand(4, 4)
plt.figure(figsize=(6, 5))
sns.heatmap(matrix, annot=True, cmap="coolwarm", xticklabels=["A", "B", "C", "D"], yticklabels=["W", "X", "Y", "Z"])
plt.title("4x4 Random Matrix Heatmap")
plt.show()

# Task 3: Generate two 4x4 matrices and perform operations
matrix1 = np.random.rand(4, 4)
matrix2 = np.random.rand(4, 4)

sum_matrix = matrix1 + matrix2
sub_matrix = matrix1 - matrix2
mul_matrix = matrix1 * matrix2

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
operations = {"Addition": sum_matrix, "Subtraction": sub_matrix, "Multiplication": mul_matrix}

for ax, (op, mat) in zip(axes, operations.items()):
    sns.heatmap(mat, annot=True, cmap="viridis", ax=ax)
    ax.set_title(op)

plt.show()
```
